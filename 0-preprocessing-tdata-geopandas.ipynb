```python
from datetime import datetime
import pandas as pd
import geopandas as gpd
```

```python
# Predefined variable for title and version
date = datetime.now()
# Enter
initials = "nm"
site = "tongatapu"

version = f"{initials}-{site}-{date.strftime('%d%m%Y')}"
print(version)
```

```python
# Read reference tdata with correct schema 
tdata = gpd.read_file(f"training-data/{version}.geojson")
```

```python
tdata = gpd.read_file("training-data/nm-tongatapu-11122025_postcard_4.geojson")
```

```python
print('list of', tdata['LULC_class'].nunique(),'unique classes')
print(tdata['LULC_class'].unique())
```

```python
print(tdata['LULC_class'].value_counts())
print('total gps points',(len(tdata)))
```

```python
tdata.loc[tdata['LULC_class'] == 'Shrubs', 'LULC_class'] = 'Grassland'

tdata.loc[tdata['LULC_class'] == 'Shallow_ocean', 'LULC_class'] = 'Water'
tdata.loc[tdata['LULC_class'] == 'Coral_reef', 'LULC_class'] = 'Water'
tdata.loc[tdata['LULC_class'] == 'Seagrass', 'LULC_class'] = 'Water'
tdata.loc[tdata['LULC_class'] == 'Deep_ocean', 'LULC_class'] = 'Water'
tdata.loc[tdata['LULC_class'] == 'Surface_water', 'LULC_class'] = 'Water'
```

```python
tdata.loc[tdata['LULC_class'] == 'Grassland', 'LULC_class'] = 'Grazing_Cropland'
tdata.loc[tdata['LULC_class'] == 'Cropland', 'LULC_class'] = 'Grazing_Cropland'
```

```python
# Define two lists
old_categories = ['Forest_land', 'Grazing_Cropland', 'Grazing_Cropland', 'Wetland', 'Settlements', 'Bare_land', 'Water']
new_groups = [1, 2, 2, 3, 4, 5, 6]

# Create the mapping dictionary
category_mapping = dict(zip(old_categories, new_groups))

# Apply mapping
tdata['LULC_code'] = tdata['LULC_class'].map(category_mapping)
```

```python
print(tdata['LULC_code'].value_counts())
print('total gps points',(len(tdata)))
```

```python
# 1) Show current cross-tab of class -> codes to inspect inconsistency
print("Cross-tab of LULC_class vs LULC_code (counts):")
ct = pd.crosstab(tdata['LULC_class'], tdata['LULC_code'])
display(ct)
```
